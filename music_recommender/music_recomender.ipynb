```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

music = pd.read_csv('music.csv') 
X = music.drop(columns = 'genre')  
y = music['genre']
model = DecisionTreeClassifier()
model.fit(X, y) #train the model on music 
prediction = model.predict([[21, 1]])   #lets predict age 21 male 
print(prediction)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# lets split the data into training and test set to check the model accuray of predicttion
#adding random state makes sure each time it runs it will be thesame value for the trian and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 2) 
model.fit(X_train, y_train) #train model on test set
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)
print(score)

#the dataset is a small one hence the prediction is almost accurate 
```
